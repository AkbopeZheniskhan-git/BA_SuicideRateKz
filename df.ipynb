```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the CSV file
data = pd.read_csv("master.csv")

# Select the top 5 countries with the highest number of suicides
top_5_countries = data.groupby("country")["suicides_no"].sum().nlargest(5).index.tolist()

# Append Kazakhstan to the list
top_5_countries.append("Kazakhstan")

# Filter the data for Kazakhstan and the top 5 countries
kazakhstan_data = data[data["country"] == "Kazakhstan"]
top_5_countries_data = data[data["country"].isin(top_5_countries)]

# Create a bar chart showing the number of suicides per country
plt.figure(figsize=(10, 6))
plt.bar(top_5_countries_data["country"], top_5_countries_data["suicides_no"])
plt.xlabel("Country")
plt.ylabel("Number of Suicides")
plt.title("Comparison of Suicides in Kazakhstan and Top 5 Countries")
plt.grid(True)
plt.show()
```

```python
import statsmodels.api as sm
# Read the data from the CSV file
data = pd.read_csv("master.csv")

# Filter the data for Kazakhstan
kazakhstan_data = data[data["country"] == "Kazakhstan"]

# Create a scatter plot showing the relationship between suicides and GDP per capita
plt.figure(figsize=(10, 6))
plt.scatter(kazakhstan_data["gdp_per_capita ($)"], kazakhstan_data["suicides_no"])
plt.xlabel("GDP per Capita ($)")
plt.ylabel("Number of Suicides")
plt.title("Relationship between Suicides and GDP per Capita in Kazakhstan")
plt.grid(True)
plt.show()

# Fit a linear regression model to the data
import statsmodels.api as sm

model = sm.OLS(kazakhstan_data["suicides_no"], kazakhstan_data["gdp_per_capita ($)"])
results = model.fit()

# Print the regression coefficients
print(results.params)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the CSV file
data = pd.read_csv("master.csv")

# Calculate the suicide rate per 100,000 people by sex
sex_grouped_suicide_rates = data.groupby("sex")["suicides/100k pop"].mean()

# Create a bar chart showing the suicide rate per 100,000 people by sex
plt.figure(figsize=(10, 6))
plt.bar(sex_grouped_suicide_rates.index, sex_grouped_suicide_rates.values)
plt.xlabel("Sex")
plt.ylabel("Suicide Rate per 100,000 People")
plt.title("Suicide Rate per 100,000 People by Sex")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Ð’Ñ‹Ð±ÐµÑ€Ð¸Ñ‚Ðµ Ð´Ð°Ð½Ð½Ñ‹Ðµ Ð´Ð»Ñ ÐšÐ°Ð·Ð°Ñ…ÑÑ‚Ð°Ð½Ð° Ð¸ Ð´Ñ€ÑƒÐ³Ð¸Ñ… ÑÑ‚Ñ€Ð°Ð½ (Ð·Ð°Ð¼ÐµÐ½Ð¸Ñ‚Ðµ 'Another Country' Ð½Ð° Ñ„Ð°ÐºÑ‚Ð¸Ñ‡ÐµÑÐºÐ¾Ðµ Ð¸Ð¼Ñ)
countries_of_interest = ['Kazakhstan', 'Uzbekistan', 'Kyrgyzstan', 'Turkmenistan']
df_countries = df[df['country'].isin(countries_of_interest)]

# Ð£Ð´Ð°Ð»Ð¸Ñ‚Ðµ ÑÑ‚Ñ€Ð¾ÐºÐ¸ Ñ Ð¿Ñ€Ð¾Ð¿ÑƒÑ‰ÐµÐ½Ð½Ñ‹Ð¼Ð¸ Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸ÑÐ¼Ð¸
df_countries = df_countries.dropna(subset=['suicides/100k pop', 'year'])

# Ð“Ñ€ÑƒÐ¿Ð¿Ð¸Ñ€Ð¾Ð²ÐºÐ° Ð´Ð°Ð½Ð½Ñ‹Ñ… Ð¿Ð¾ ÑÑ‚Ñ€Ð°Ð½Ðµ Ð¸ Ð³Ð¾Ð´Ð°Ð¼, Ð° Ð·Ð°Ñ‚ÐµÐ¼ Ð¿Ð¾Ð´ÑÑ‡ÐµÑ‚ ÑÑ€ÐµÐ´Ð½ÐµÐ³Ð¾ ÑƒÑ€Ð¾Ð²Ð½Ñ ÑÐ°Ð¼Ð¾ÑƒÐ±Ð¸Ð¹ÑÑ‚Ð² Ð½Ð° 100 Ñ‚Ñ‹Ñ. Ñ‡ÐµÐ»Ð¾Ð²ÐµÐº
average_suicides_by_country_year = df_countries.groupby(['country', 'year'])['suicides/100k pop'].mean().reset_index()

# Ð’Ð¸Ð·ÑƒÐ°Ð»Ð¸Ð·Ð°Ñ†Ð¸Ñ ÑÑ€Ð°Ð²Ð½Ð¸Ñ‚ÐµÐ»ÑŒÐ½Ð¾Ð³Ð¾ Ð»Ð¸Ð½ÐµÐ¹Ð½Ð¾Ð³Ð¾ Ð³Ñ€Ð°Ñ„Ð¸ÐºÐ°
plt.figure(figsize=(12, 8))
sns.lineplot(x='year', y='suicides/100k pop', hue='country', data=average_suicides_by_country_year)
plt.title('Comparison of Suicide Rates (per 100k) between Countries Over Years')
plt.xlabel('Year')
plt.ylabel('Average Suicides/100k pop')
plt.show()
```
